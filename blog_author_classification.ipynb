###Get counts of authors in blog

```python
author_to_counts=Counter()
for text in blog.find():
    author_to_counts[text['author']]+=1
print author_to_counts
print '-------------------'
minimum_threshold=sum(author_to_counts.values())*.05
common_author_list=[author for author,count in author_to_counts.iteritems() 
                    if count > minimum_threshold and author.count('.')>=2]    
print common_author_list
```

```python
blog_list=[]
for post in blog.find():
    #don't want short ones
    if len(post['text'])>1500:
        if post['author'] in common_author_list:
            author=post['author']
        else:
            author='Other'
        blog_list.append((author,post['text']))
```

```python
author_list=[]
blog_text_list=[]
shuffle(blog_list)
for i in blog_list:
    author_list.append(i[0])
    blog_text_list.append(i[1])
```

```python
pd.Series(author_list).value_counts(normalize=False, ascending=True).plot(kind='barh')
plt.title("Counts of authors in blog posts")
plt.xlabel("Frequency")
plt.ylabel("Author")
```

### Logisitic Regression

```python
kf = cross_validation.KFold(len(author_list), n_folds=5)
X_train, X_test, y_train, y_test = cross_validation.train_test_split(blog_text_list,author_list)
vectorizer = TfidfVectorizer(max_features=3000,max_df=.7,min_df=10,ngram_range=(1,4))

for i in np.arange(.01,1,.1):
    logreg = linear_model.LogisticRegression(C=i)
    doc_vectors = vectorizer.fit_transform(X_train)
    logreg.fit(doc_vectors,y_train)
    test_vector = vectorizer.transform(X_test)
    y_pred=logreg.predict(test_vector)
    print accuracy_score(y_test,y_pred)
```

###Bernoulli Naive Bayes classifier

```python
kf = cross_validation.KFold(len(author_list), n_folds=5)
X_train, X_test, y_train, y_test = cross_validation.train_test_split(blog_text_list,author_list)
vectorizer = TfidfVectorizer(max_df=.7,min_df=10,ngram_range=(1,3))
doc_vectors = vectorizer.fit_transform(X_train)
print doc_vectors.shape
model = BernoulliNB().fit(doc_vectors,y_train)
test_vector = vectorizer.transform(X_test)
y_pred=model.predict(test_vector)
print classification_report(y_test, y_pred)
print accuracy_score(y_test,y_pred)
```

```python
top_words_dict_with_values={}
for j in range(5):
    top_indices=np.argsort(model.feature_log_prob_[j]-np.mean(model.feature_log_prob_, axis=0))[::-1][:20]
    top_values=(model.feature_log_prob_[j]-np.mean(model.feature_log_prob_, axis=0))[top_indices]
    top_words=list(np.asarray(vectorizer.get_feature_names())[top_indices])
    top_words_dict_with_values[model.classes_[j]]=dict(zip(top_words,top_values))
```
