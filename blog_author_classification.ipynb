```python
#make dict with most popular 150 words
top_words_dict={}
for j in range(5):
    print 
#     print model.classes_[j]
    top_30_words=np.argsort(model.feature_log_prob_[j]-np.mean(model.feature_log_prob_, axis=0))[::-1][:150]
#     print np.asarray(vectorizer.get_feature_names())[top_30_words]
    top_words_dict[model.classes_[j]]=list(np.asarray(vectorizer.get_feature_names())[top_30_words])
```

```python
top_words_dict_with_values={}
for j in range(5):
    print 
#     print model.classes_[j]
    top_indices=np.argsort(model.feature_log_prob_[j]-np.mean(model.feature_log_prob_, axis=0))[::-1][:20]
    top_values=(model.feature_log_prob_[j]-np.mean(model.feature_log_prob_, axis=0))[top_indices]
    top_words=list(np.asarray(vectorizer.get_feature_names())[top_indices])
    #print top_words
    top_words_dict_with_values[model.classes_[j]]=dict(zip(top_words,top_values))
```

```python
from textblob import TextBlob

counts_list=[]
for author,corpus in article_texts.iteritems():
    for text in corpus:
        text_dict={}
        text_dict['author']=author
        text_dict['text']=text
        text_dict['word_count']={}
        blob = TextBlob(text_dict['text'])
        for top_word in top_words_dict[author]:
            #print top_word
            #print type(text_dict['text'])
            n = len(top_word.split())
            if n == 1 and top_word in blob.words:
                # We have a single word top-word in here
                #print 'yes', top_word
                text_dict['word_count'][top_word] = blob.words.count(top_word)
            elif n > 1:
                top_n_gram = top_word.split()
                all_text_ngrams = blob.ngrams(n)
                n_gram_count = all_text_ngrams.count(top_n_gram)
                if n_gram_count > 0:
                    #print 'yes', top_word
                    text_dict['word_count'][top_word] = n_gram_count

        counts_list.append(text_dict)
```

```python
def plot_confidence(model,example,y_true, axes):
    confidences = model.predict_log_proba(example)[0]
    #print y_true
    #print confidences
    true_class_index = list(model.classes_).index(y_true)
    pred_class_index = np.argmax(confidences)
    colors = ['b'] * len(model.classes_)
    colors[true_class_index] = 'm'
    if true_class_index != pred_class_index:
        colors[pred_class_index] = 'r'
    plt.bar(range(len(confidences)), confidences, color=colors)
    axes.set_xticklabels( model.classes_ )
    
    

plt.figure(figsize=(8,50))
counter = 1
for i, vector in enumerate(test_vector[10:18]):
    if model.predict(vector) != y_pred[i] or True:
        counter += 1
        ax = plt.subplot(9,1,counter)
        plot_confidence(model, vector,y_pred[i], ax )
```
